```python
with open("ALL_GPT-4_TOKENS.txt", "r") as f:
    tokens = f.read().splitlines()

numbers = [*filter(str.isnumeric, tokens)]
print(len(numbers))
```

```python
numbers
```

```python
print([f"{n:02}" for n in range(100) if f"{n:02}" not in numbers])
print([f"{n:03}" for n in range(100) if f"{n:03}" not in numbers])
print([n for n in range(1000) if str(n) not in numbers])
```
